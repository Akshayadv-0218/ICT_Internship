Outlier Detection

```python
num_columns = ['SSN','Annual_Income','Monthly_Inhand_Salary','Num_Bank_Accounts','Num_Credit_Card','Interest_Rate','Num_of_Loan','Delay_from_due_date','Num_of_Delayed_Payment','Changed_Credit_Limit','Num_Credit_Inquiries','Outstanding_Debt','Credit_Utilization_Ratio','Credit_History_Age','Total_EMI_per_month','Amount_invested_monthly','Monthly_Balance']
for i in num_columns:
    plt.figure()
    plt.boxplot(data[i])
    plt.title(i)
plt.show()
```

Label Encoding

```python
#Label encoding
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
to_encode=['Credit_Mix','Occupation','Type_of_Loan','Payment_of_Min_Amount','Payment_Behaviour','Credit_Score']
for col in to_encode:
    data[col]=le.fit_transform(data[col])
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
```

```python
num_col = ['SSN','Monthly_Balance','Annual_Income','Monthly_Inhand_Salary','Outstanding_Debt','Type_of_Loan','Total_EMI_per_month']
```
